```python
import sqlite3
import pandas as pd
conexion = sqlite3.connect("bd1.db")
consulta_sql = '''
SELECT "Cambio Anual", SUM("Proy. Exp. de la poblacion") AS Total_Población 
FROM paises_del_mundo
GROUP BY "Cambio Anual";
'''

resultado = pd.read_sql(consulta_sql, conexion)
print(resultado)
conexion.close()
```

```python
#consulta paises con proyeccion entre 800000 y 1200000
import sqlite3
import pandas as pd
conexion = sqlite3.connect("bd1.db")
consulta_sql = '''
SELECT "Pais"
FROM paises_del_mundo
WHERE "Proy. Exp. de la poblacion" BETWEEN 800000 AND 1200000 
ORDER BY "Proy. Exp. de la poblacion" ASC;
'''
resultado = pd.read_sql(consulta_sql, conexion)
print(resultado)
conexion.close()
```

```python
import shutil
import pandas as pd
import sqlite3

# Paso 1: Mover el archivo CSV a una ubicación accesible
origen = r"C:\Users\elcar\paisesinstant.csv"  # Ruta actual del archivo
destino = r"C:\Users\elcar\Documents\paisesinstant.csv"  # Nueva ubicación

# Muevo el archivo
shutil.move(origen, destino)

print(f"Archivo movido a: {destino}")

# Paso 2: Leo el archivo CSV desde la nueva ubicación
df1 = pd.read_csv(destino, sep=',', encoding='latin1')

# Paso 3: Conecto a SQLite
conn = sqlite3.connect('mi_base_de_datos.db')

# Paso 4: Guardo el DataFrame en SQLite
df1.to_sql('nombre_de_la_tabla', conn, if_exists='replace', index=False)

print("Archivo cargado en SQLite con éxito.")

# Paso 5: Valido la carga de datos
query_result = pd.read_sql('SELECT * FROM nombre_de_la_tabla LIMIT 5', conn)
print(query_result)


conn.close()
```
